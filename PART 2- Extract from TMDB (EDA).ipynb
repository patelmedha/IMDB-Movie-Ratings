# Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import csv
#Folder path must end with /
FOLDER = "Data/"
```

## Load in your csv.gz's of results for each year extracted.


```python
movies_00 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
movies_00.head()
```

```python
movies_01 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
movies_01.head()
```

## Concatenate the data into 1 dataframe for the remainder of the analysis.

```python
movies00_01 = pd.concat([movies_00, movies_01], ignore_index=True)
movies00_01.head()
```

```python
movies00_01.info()
```

1. How many movies had at least some valid financial information (values > 0 for budget OR revenue)?
    - Please exclude any movies with 0's for budget AND revenue from the remaining visualizations.


```python
fin_valid = movies00_01[(movies00_01['revenue'] > 0) | (movies00_01['budget'] > 0)]
fin_valid = fin_valid['imdb_id'].value_counts().sum()
print(f'There are {fin_valid} movies with valid financial information.')
```

```python
# Plotting the information using a bar plot
plt.figure(figsize=(6, 4))
plt.bar(['Movies with Valid Financial Info (2000-2001)'], [fin_valid], color='skyblue')
plt.title('Number of Movies with Valid Financial Information (2000-2001)')
plt.ylabel('Count')

# Saving the plot as an image file
plt.savefig('Images/valid_financial_movies00_01')

# Show the plot
plt.show()
```

2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?


```python
certification_count = movies00_01['certification'].value_counts()
print(f'Total number of movies per certification category is:')
print(f'{certification_count}')
```

```python
fig , ax = plt.subplots()

ax= sns.histplot(data = movies00_01, x = 'certification')

#Set title name and axis names
ax.set_title('Movies per Certification Category (2000-2001)')
ax.set_xlabel('Certification')
ax.set_ylabel("Count");

# Saving plot to images folder
plt.savefig('Images/movies00_01_per_category')
```

3. What is the average revenue per certification category?

```python
avg_revenue_per_certi = movies00_01.groupby('certification')['revenue'].mean().round()
print(f'The average revenue per certification category is:')
print(f'\n')
print(f'{avg_revenue_per_certi}')
```

4. What is the average budget per certification category?


```python
avg_budget_per_certi = movies00_01.groupby('certification')['budget'].mean().round()
print(f'The average budget per certification category is:')
print(f'\n')
print(f'{avg_budget_per_certi}')
```

```python
## Define label_order
avg_budget_per_certi = movies00_01.groupby('certification')['budget'].mean().sort_values(ascending=False)
avg_budget_per_certi
## Define a MatplotLib ax object using sns.barplot()
fig, ax = plt.subplots()
ax = sns.barplot(data = movies00_01,
                 x ='certification',
                 y = 'budget',
                 order = avg_budget_per_certi.index,
                 errorbar = None)
## Set the Title
ax.set_title('Movies Average Budget by Certification (2000-2001)', fontsize=12, fontweight = 'bold')
# Set Axes Labels
ax.set_xlabel('Movie Certification', fontsize=10, fontweight = 'bold')
ax.set_ylabel('Budget(USD)', fontsize=10, fontweight = 'bold');

#Saving plot
plt.savefig('Images/avg_budget_certification_movies00_01.png')
```
